```python
x1 = np.linspace(-9,9,100)
x2 = np.linspace(-1,1,5)
# 頭頂
y1 = x1*x1*np.cos(x1)+ 77.30155121
# 下巴
y2 = x1**2/2-160
# 嘴巴
y3 = -(x2*1.5)**2-80
# 畫頭頂下巴嘴巴
plt.plot(x1,y1,"k",x1,y2,"k",x2,y3,"k")
# 畫鼻子
plt.plot(np.cos(x1),10*np.sin(x1),"k")
# 畫右眼
plt.plot(np.cos(x1)+5,15*np.sin(x1)+25,"k")
plt.plot(1.2*np.cos(x1)+5,20*np.sin(x1)+25,"k")
# 畫左眼
plt.plot(np.cos(x1)-5,15*np.sin(x1)+25,"k")
plt.plot(1.2*np.cos(x1)-5,20*np.sin(x1)+25,"k")
# 畫左鬍鬚
plt.hlines(-25,-9,-7,"r")
plt.hlines(-25,7,9,"r")
plt.text(-2,100,"Facepalm")
plt.show()
```
