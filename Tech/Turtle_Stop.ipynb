> [Stop]!(Stop.JPG)
> 不用Cycle，建置一個函式(def)

## -注意事項
> 測試時發現turtle的模組內'無法讀取字串，需先轉乘"，故str().replace("'", "\"")

```python
import turtle
import numpy as np
"""
亦可不用迴圈改用circle step方式寫，目前沒想到如何記錄位置
可變更參數
color_of_the_pen 筆的顏色
fill_of_the_pen 填上的色彩
text_of_the_word 字的顏色
differ_change 內外圈差
text_size 字大小
"""
color_of_the_pen = "red"
fill_of_the_pen = "red"
text_of_the_word = "white"
differ_change = 20
exterior_path = 100
text_size = 40


def cycle(path, angle=60):
    """
    path為前進的步伐
    angle為右轉的角度
    location_place 位置紀錄被儲存
    cycle_place 內圈初始位置
    word_place 文字位置

    """
    global cycle_place, word_place
    i = 0
    location_place = []

    while i < 6:
        # 紀錄外圈每次烏龜走的位置
        location_place.append(turtle.pos())
        i += 1
        # 右轉60度
        turtle.right(angle)
        # 向前100
        turtle.forward(path)
    # 用numpy轉成可計算
    location_place = np.array(location_place)
    # 拿起筆
    turtle.penup()
    # 內圈初始位置
    cycle_place = (location_place[1][0]-differ_change, location_place[1][1])
    # 走至內圈起始位置
    turtle.goto(cycle_place)
    # 放下筆
    turtle.pendown()
    # 計算文字位置
    word_place = ((location_place[1][0]+location_place[4][0])/2,
                  (location_place[1][1]+location_place[4][1])/2 -
                   1.5 * differ_change)
    return word_place


# -主程式
# 將輸出的單引號轉成"才能放進turtle模組(Module)的方法(Method)
color_of_the_pen = str(color_of_the_pen).replace("'", "\"")
fill_of_the_pen = str(fill_of_the_pen).replace("'", "\"")
text_of_the_word = str(text_of_the_word).replace("'", "\"")

# 變更筆顏色
turtle.pencolor(color_of_the_pen)
# 變更筆粗度
turtle.pensize(10)
# 執行外圈
cycle(exterior_path)
# 顯示烏龜的標示
turtle.showturtle()
# 旋轉烏龜初始位置的角度
turtle.right(60)
# 執行內圈
# 計算內圈大小
interior_path = exterior_path - differ_change
# 執行內圈
# 變更內圈顏色
turtle.color(color_of_the_pen, fill_of_the_pen)
# 開始填色
turtle.begin_fill()
# 呼叫函式cycle
cycle(interior_path)
# 結束填色
turtle.end_fill()
# 執行文字
turtle.pencolor(text_of_the_word)
turtle.penup()
turtle.goto(word_place)
turtle.pendown()
turtle.write("STOP", align="center", font=(
    "Arial", text_size, "normal"))
turtle.hideturtle()
```
